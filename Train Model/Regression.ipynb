```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model 
import math
from pathlib import Path
```

```python
input_data_name = "Clean_Data.csv"
script_dir = Path().absolute().parent            # Get the current script's directory
input_data_path = script_dir / "Resources" / "Text" / input_data_name
```

```python
df = pd.read_csv(input_data_path)
df.plot(kind = 'scatter', x = 'y', y = 'delta_x')
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['y']], df.delta_x)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.xlabel('y')
plt.ylabel('delta_x')
plt.scatter(df.y, df.delta_x, marker='+')
plt.plot(df.y, reg.predict(df[['y']]), color = 'red')
```

```python
k  = -reg.coef_[0]
mu = reg.intercept_

Y = df['y'].values
deltaX = df['delta_x'].values
squares = [i for i in range(len(Y))]
sum = 0


for i in range(len(Y)):
    square = (deltaX[i] + k*(Y[i]) - mu)**2
    squares[i] = square
    sum = sum + square

dataframe = {
    'y' : Y,
    'deltaX' : deltaX,

    '(deltaX + ky - mu)^2' : squares
}

sigma = math.sqrt(sum / (len(Y)))

print(f"k = {k}")
print(f"mu = {mu}")
print(f"sigma = {sigma}")
```
